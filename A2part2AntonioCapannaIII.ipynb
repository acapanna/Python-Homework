```
X = df.drop(['Population2010', 'Population2010Binary'], axis=1)
Y = df['Population2010Binary']
```

```
array = df.values
X = X.dropna()
Y = Y.dropna()
validation_size = 0.20
seed = 123
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=validation_size, random_state=seed)

# Evaluate Algorithms
# Test options and evaluation metric
num_folds = 20
seed = 123
scoring = 'neg_mean_squared_error'
```

```
#instead of Spotcheck, use the Pipeline example
pipelines = []
pipelines.append(('ScaledBOOST', Pipeline([('Scaler', StandardScaler()),('BOOST', GradientBoostingClassifier())])))
pipelines.append(('ScaledCCLASS', Pipeline([('Scaler', StandardScaler()),('CCLASS', DecisionTreeClassifier())])))
pipelines.append(('ScaledFOREST', Pipeline([('Scaler', StandardScaler()),('FOREST', RandomForestClassifier())])))
pipelines.append(('ScaledLDA', Pipeline([('Scaler', StandardScaler()),('LDA', LinearDiscriminantAnalysis())])))
pipelines.append(('ScaledLOG', Pipeline([('Scaler', StandardScaler()),('LOG', LogisticRegression(max_iter=1000000))])))
pipelines.append(('ScaledKNNCLASS', Pipeline([('Scaler', StandardScaler()),('KNNCLASS', KNeighborsClassifier())])))
pipelines.append(('ScaledGAUS', Pipeline([('Scaler', StandardScaler()),('GAUS', GaussianNB())])))
pipelines.append(('ScaledETREES', Pipeline([('Scaler', StandardScaler()),('ETREES', ExtraTreesClassifier())])))
pipelines.append(('ScaledBAGS', Pipeline([('Scaler', StandardScaler()),('BAGS', BaggingClassifier())])))
results = []
names = []
for name, model in pipelines:
	kfold = KFold(n_splits=num_folds, random_state=seed,
               shuffle=True)
	cv_results = cross_val_score(model, X_train, Y_train, 
                              cv=kfold, 
                              scoring='accuracy')
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
```

```
#PIPELINES FOR BOOST, DECISION TREE, AND GAUSNB
#data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=validation_size, random_state=seed)

#Build the pipeline
pipe_boost = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=2)),
			('clf', GradientBoostingClassifier(random_state=seed))])

pipe_cclass = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=2)),
			('clf', DecisionTreeClassifier(random_state=seed))])

pipe_gaus = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=2)),
			('clf', GaussianNB(priors=None))])

param_range = [1, 2, 3, 4]

# Set grid search params
grid_params_boost = [{'clf__criterion': ['friedman_mse'],
		'clf__min_samples_leaf': param_range,
		'clf__max_depth': param_range,
		'clf__min_samples_split': param_range[1:]}]

grid_params_cclass = [{'clf__criterion': ['gini', 'entropy'],
		'clf__min_samples_leaf': param_range,
		'clf__max_depth': param_range,
		'clf__min_samples_split': param_range[1:]}]

grid_params_gaus = [{}]

#construct grid search(when I first ran it kfold had to be 20 but now it takes kfold. Don't really know why.)
# Construct grid searches
jobs = -1

gs_boost = GridSearchCV(estimator=pipe_boost,
			param_grid=grid_params_boost,
			scoring='accuracy',
			cv=kfold,
      n_jobs=jobs)
			
gs_cclass = GridSearchCV(estimator=pipe_cclass,
			param_grid=grid_params_cclass,
			scoring='accuracy',
			cv=kfold, 
			n_jobs=jobs)

gs_gaus = GridSearchCV(estimator=pipe_gaus,
			param_grid=grid_params_gaus,
			scoring='accuracy',
			cv=kfold)

# List of pipelines for ease of iteration
grids = [gs_boost, gs_cclass, gs_gaus]

# Dictionary of pipelines and classifier types for ease of reference
grid_dict = {0: 'Gradient Boosting Classifier', 1: 'Decision Tree Classifier',
             2: 'GaussianNB'}

# Fit the grid search objects
print('Performing model optimizations...')
best_acc = 0.0
best_clf = 0
best_gs = ''
for idx, gs in enumerate(grids):
	print('\nEstimator: %s' % grid_dict[idx])	
	# Fit grid search	
	gs.fit(X_train, Y_train)
	# Best params
	print('Best params: %s' % gs.best_params_)
	# Best training data accuracy
	print('Best training accuracy: %.3f' % gs.best_score_)
	# Predict on test data with best params
	y_pred = gs.predict(X_test)
	# Test data accuracy of model with best params
	print('Test set accuracy score for best params: %.3f ' % accuracy_score(Y_test, y_pred))
 	# Track best (highest test accuracy) model
	if accuracy_score(Y_test, y_pred) > best_acc:
		best_acc = accuracy_score(Y_test, y_pred)
		best_gs = gs
		best_clf = idx
print('\nClassifier with best test set accuracy: %s' % grid_dict[best_clf])
```
